```python
import gradio as gr
import uuid
from app.chat.chat_service import process_query
from app.database.manual import save_manual_knowledge

def chat(message, session_id):

    # Generate unique session per browser session
    if session_id is None:
        session_id = str(uuid.uuid4())

    result = process_query(message, debug=True, session_id=session_id)

    return (
        result["answer"],
        result["retrieved_docs"],
        message,
        result["answer"],
        session_id,
        result["source"]
    )

def save_teach(question, answer):
    if question.strip() == "" or answer.strip() == "":
        return "Question and Answer cannot be empty."
    
    save_manual_knowledge(question, answer)
    return "Saved to manual knowledge base!"

with gr.Blocks() as demo:

    # IMPORTANT: start as None (NOT create_session())
    session_state = gr.State(None)

    gr.Markdown("# IDEAM Dev Chatbot")

    input_box = gr.Textbox(label="Ask IDEAM something")
    answer_box = gr.Textbox(label="Answer")
    retrieved_box = gr.JSON(label="Retrieved Docs")
    source_box = gr.Textbox(label="Answer Source")

    gr.Markdown("## Teach Me Section")

    teach_question = gr.Textbox(label="Correct Question")
    teach_answer = gr.Textbox(label="Correct Answer")
    teach_button = gr.Button("Save Knowledge")
    teach_status = gr.Textbox(label="Status")

    input_box.submit(
        chat,
        [input_box, session_state],
        [answer_box, retrieved_box, teach_question, teach_answer, session_state, source_box]
    )

    teach_button.click(
        save_teach,
        [teach_question, teach_answer],
        teach_status
    )

demo.launch(share=True)
```

```python
import psycopg2
import pandas as pd
import os
from datetime import datetime

# ===== CONFIG =====
DB_NAME = "chatbot_test"
EXPORT_FOLDER = "data/exports"

# Create export folder if not exists
os.makedirs(EXPORT_FOLDER, exist_ok=True)

# Timestamp for file naming
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Connect to DB
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database=DB_NAME,
    user="postgres",
    password="postgres"
)

tables = [
    "knowledge_chunks",
    "manual_knowledge",
    "chat_history",
    "review_questions"
]

for table in tables:
    query = f"SELECT * FROM {table};"
    df = pd.read_sql(query, conn)
    
    file_path = os.path.join(
        EXPORT_FOLDER,
        f"{table}_{timestamp}.csv"
    )
    
    df.to_csv(file_path, index=False)
    print(f"✅ Exported {table} → {file_path}")

conn.close()

print("\n All tables exported successfully.")
```
